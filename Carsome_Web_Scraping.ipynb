```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
baseurl = 'https://www.carsome.id'
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36 Edg/123.0.2420.81'}
```

```python
# Mendapatkan seluruh link mobil dari semua page carsome

productlinks = []
for x in range(1,8):
  r=requests.get(f'https://www.carsome.id/beli-mobil-bekas?pageNo={x}', headers=headers)

  soup = BeautifulSoup(r.content, 'lxml')

  productlist = soup.find_all('div', class_='list-card__item')

  for item in productlist:
    for link in item.find_all('a', href=True):
      productlinks.append(baseurl + link['href'])

print(len(productlinks))
print(productlinks)
```

```python
# Mengakses dan meng-scrap data dari setiap link yang ada

carlist= []
for link in productlinks:

    r = requests.get(link, headers=headers)

    soup = BeautifulSoup(r.content, 'lxml')

    nama = soup.find('div', class_='vehicle__title-wrapper car-title-wrapper').text.strip()
    jarak_tempuh = soup.find('div', class_='car-mileage').text.strip()
    harga = soup.find('div', class_='price').text.strip()
    bahan_bakar = soup.find('span', class_='value').text.strip()
    warna = soup.find_all('span', class_='value')[1].text.strip()
    kursi = soup.find_all('span', class_='value')[2].text.strip()
    garansi_pabrik = soup.find_all('span', class_='value')[8].text.strip()
    tgl_registrasi = soup.find_all('span', class_='value')[3].text.strip()
    tipe_registrasi = soup.find_all('span', class_='value')[4].text.strip()
    kunci_cadangan = soup.find_all('span', class_='value')[6].text.strip()
    buku_service = soup.find_all('span', class_='value')[7].text.strip()
    STNK = soup.find_all('span', class_='value')[9].text.strip()

    car = {
        'nama' : nama,
        'jarak_tempuh' : jarak_tempuh,
        'harga' : harga,
        'bahan_bakar' : bahan_bakar,
        'warna' : warna,
        'kursi' : kursi,
        'garansi_pabrik' : garansi_pabrik,
        'tgl_registrasi' : tgl_registrasi,
        'tipe_registrasi' : tipe_registrasi,
        'kunci_cadangan' : kunci_cadangan,
        'buku_service' : buku_service,
        'STNK' : STNK
    }
    carlist.append(car)

    #print(car)
```

```python
# Menyimpan data dalam bentuk dataframe
df = pd.DataFrame(carlist)
```
